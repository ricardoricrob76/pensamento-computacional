```python
# import das bibliotecas de Machine Learning
from sklearn.datasets import load_iris # carregar a base de dados iris.dataset
from sklearn.model_selection import train_test_split # separa a base 80% treino e 20% teste
from sklearn.ensemble import RandomForestClassifier # Algoritmo de ML
from sklearn.metrics import accuracy_score, classification_report # Avaliar Modelo Machine Learning
import pandas as pd # principal lib para manipulação de dados (Excel)
```

```python
# Carregando o dataset Íris
dados = load_iris()
caracteristicas = dados.data  # Características (comprimento/largura das pétalas e sépalas)
classes = dados.target  # Classes das flores
```

```python
# Dividir os dados em conjunto de treino e teste
caracteristicas_treino, caracteristicas_teste, classes_treino, classes_teste = train_test_split(caracteristicas, classes, test_size=0.35, random_state=42)
```

```python
# Criar e treinar o modelo Random Forest
modelo = RandomForestClassifier(n_estimators=100, random_state=42)
modelo.fit(caracteristicas_treino, classes_treino)
```

```python
# Fazer previsões
previsoes = modelo.predict(caracteristicas_teste)
```

```python
# Avaliar o modelo
acuracia = accuracy_score(classes_teste, previsoes)
print(f'Acurácia: {acuracia:.2f}')
print('Relatório de Classificação:')
print(classification_report(classes_teste, previsoes, target_names=dados.target_names))
```

```python
# A acurácia deu 100%, o que significa que houve overfitting (os dados foram decorados, e não aprendidos). Uma possível solução seria
# aumentar a quantidade de dados para treino.
```
